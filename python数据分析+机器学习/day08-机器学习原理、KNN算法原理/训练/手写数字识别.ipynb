```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
# 魔法函数，不能再pycharm使用，可以在jupyter中使用
# 可以在IPython编译器中直接使用，功能可以内嵌绘图，省略掉plt.show()这一步
%matplotlib inline
```

```python
train=[]
target=[]

# 读取文件
for j in range(10):
#     display(j)  ./0/0_1.bmp
    for i in range(1,501):
        img_path='./%d/%d_%d.bmp'%(j,j,i)
#         读取图片数据并进行扁平化处理
        img_data=plt.imread(img_path).ravel()
        train.append(img_data)
#         每次获取一个图片的数据，就把target保存起来
        target.append(j)
        
```

```python
# 数据切分
from sklearn.model_selection import train_test_split
#取75%数据作为训练数据    25%数据作为测试数据
X_train,X_test,y_train,y_test=train_test_split(train,target)
```

```python
display(len(X_train),len(X_test))
```

```python
display(len(y_train),len(y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
# 构建KNN对象
knnclf=KNeighborsClassifier(n_neighbors=5)
# 训练数据
knnclf.fit(X_train,y_train)
```

```python
# 预测结果
y_=knnclf.predict(X_test)
```

```python
knnclf.score(X_test,y_test)
```

```python
# 绘制图形   对比下前100条数据识别出来的值跟原来真实的值
plt.figure(figsize=(12,16))
for i in range(100):
    axes=plt.subplot(10,10,i+1)
#     cmap='gray’灰度化处理
    plt.imshow(X_test[i].reshape(28,28),cmap='gray')
#     原来真实的值
    t=y_test[i]
#     识别出来的值
    p=y_[i]
    title='T:'+str(t)+'\nP:'+str(p)
    axes.set_title(title)
    axes.axis('off')
```

```python
# 原图
plt.imshow(X_test[0].reshape(28,28),cmap='gray')
# y_test[0]  原来的数字    y_[0]：识别出来的数字
display(y_test[0],y_[0])
```
