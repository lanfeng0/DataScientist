```python
import numpy as np
# import scipy.ndimage as ndimage

import matplotlib.pyplot as plt

import sklearn.datasets as datasets

import logging as logging


# 当我们新配置的环境，第一次下载数据，半个小时，显示，加载百分之多少
logging.basicConfig(level = logging.WARN,format = '%(asctime)s %(message)s')
```

```python
face = datasets.fetch_lfw_people(min_faces_per_person=70,resize=1,
                                 slice_=(slice(70,195,None),slice(78,172,None)))
# lfw label face wild 图片进行标记
```

```python
# 有的人图片比较少
face['data'].shape
```

```python
data = face['data']

target = face['target']

target_names = face['target_names']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 保留了50张图片作为预测的数据
X_train,X_test ,y_train,y_test = train_test_split(data,target,test_size = 0.02)
```

从人脸数据中提取进行机器学习的关键数据

```python
from sklearn.decomposition import PCA
```

```python
# 62500 --->300 
# whiten作用主要正太分布式的处理，标准偏差一致化
pca = PCA(n_components=150,whiten=True)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
```

```python
X_train.shape
```

```python
X_train_pca.shape
```

```python
X_test_pca = pca.transform(X_test)
```

```python
X_test_pca.shape
```

分割训练数据和预测数据

使用PCA进行降维

使用GridSearchCV我们找到了最佳的支持向量机方法

```python
from sklearn.svm import SVC
```

```python
# C = 1.0，gamma = 0.0033
from sklearn.model_selection import GridSearchCV
```

```python
svc = SVC()

param_grid = {'C':[0.3,0.5,0.8,1.0,1.5,1.8,2.0,3.0,10],'gamma':[0.0001,0.001,0.002,0.003,0.008,0.01,0.2,0.5,1.0]}

gCV = GridSearchCV(estimator=svc,param_grid=param_grid)

gCV.fit(X_train_pca,y_train)
```

```python
svc_best = gCV.best_estimator_
svc_best
```

```python
gCV.best_score_
```

进行数据预测
