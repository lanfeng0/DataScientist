```python
from sklearn.datasets import fetch_lfw_people
```

```python
# 获取人脸数据
faces = fetch_lfw_people(resize=0.4,min_faces_per_person=100)
```

```python
data = faces.data
target = faces.target
names = faces.target_names
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(faces.images[0],cmap='gray')
```

```python
# 降维处理，拆分数据集
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

pca = PCA(n_components=0.99)
train_pca = pca.fit_transform(data)
X_train,X_test,y_train,y_test = train_test_split(train_pca,target,test_size=0.2)
```

```python
# 使用交叉验证进行模型选择
from sklearn.model_selection import cross_val_score

# 导入所有可用的分类模型
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

knn = KNeighborsClassifier()
logistic = LogisticRegression()
svc = SVC(kernel='rbf')
tree = DecisionTreeClassifier()
```
