### 二、matplotlib

- figure对象   根画布     plt.figure()
- axes对象    子画布     axes = plt.subplot(111)
- axis属性    坐标轴     axes.axis()   plt.axis()
- xlim ylim   坐标界限    axes.set_xlim() plt.xlim()
- title属性   子画布标题  axes.set_title()  plt.title()
- xlabel ylabel 轴标题    axes.set_xlabel() plt.xlabel()
- xticks yticks 轴刻度   axes.set_xticks() plt.xticks()
- xticklabels yticklabels 轴刻度标签  axes.set_xticklabels() plt.xticklabels()

### 三、 matplotlib四图绘制

- 线型图
plt.plot(x,y)

- 条形图
plt.bar(x,height,width)  x就是索引  height就是数据值  width就是条形的宽度 
plt.barh(y,width,height) y就是索引  width就是数据值  height就是条形的高度

- 直方图
plt.hist(x,bins)  x数据  bins数据分区个数

- 散点图
plt.scatter(x,y)

- 饼图
plt.scatter(x)  x是一个列表，如果列表所有数据的和不足1，就是不完整饼图
    > 饼图的外观设置参数
    - labels参数设置每一块的标签；
    - labeldistance参数设置标签距离圆心的距离（比例值,只能设置一个浮点小数）
    - autopct参数设置比例值的显示格式(%1.1f%%)；
    - pctdistance参数设置比例值文字距离圆心的距离
    - explode参数设置每一块顶点距圆形的长度（比例值,列表）；
    - colors参数设置每一块的颜色（列表）；
    - shadow参数为布尔值，设置是否绘制阴影
    - startangle参数设置饼图起始角度


- 玫瑰图
plt.axes(polar=True) 就是一个特殊的条形图
