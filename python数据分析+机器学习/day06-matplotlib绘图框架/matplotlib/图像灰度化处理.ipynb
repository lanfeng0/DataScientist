```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

# 
image = mpimg.imread('.jpeg')
print(image.shape)
# 
gray_max = np.max(image, axis=2)
print(gray_max.shape)
# 
weights = [0.2989, 0.587, 0.114]  # 
gray_weighted_avg = np.average(image, axis=2, weights=weights)
print(gray_weighted_avg.shape)
# 
gray_avg = np.mean(image, axis=2)

# 
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].imshow(image)
axes[0, 0].set_title('Original')

axes[0, 1].imshow(gray_max, cmap='gray')
axes[0, 1].set_title('Max Value Method')

axes[1, 0].imshow(gray_weighted_avg, cmap='gray')
axes[1, 0].set_title('Weighted Average Method')

axes[1, 1].imshow(gray_avg, cmap='gray')
axes[1, 1].set_title('Average Method')

for ax in axes.flat:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
